# 1) Series — crear y operar  


**Crea una Series desde una lista y otra desde un diccionario.**

```python
import pandas as pd

# 1. Series desde una lista
nombres = ['Ana', 'Bob', 'Clara', 'Diego', 'Eva']
serie_lista = pd.Series(nombres, name='Nombres')
print("Series desde lista:")
print(serie_lista)
print()

# 2. Series desde un diccionario
datos_dict = {
    'a': 25,
    'b': 30,
    'c': 22,
    'd': 35,
    'e': 28
}
serie_dict = pd.Series(datos_dict, name='Edades')
print("Series desde diccionario:")
print(serie_dict)
```

**Accede a elementos por índice, modifica un valor y realiza una operación matemática (por ejemplo, multiplicar por 2).**

```python
import pandas as pd

# Crear Series desde diccionario
datos_dict = {
    'a': 25,
    'b': 30,
    'c': 22,
    'd': 35,
    'e': 28
}
serie_edades = pd.Series(datos_dict, name='Edades')
print("Series original:")
print(serie_edades)
print()
# Acceder por etiqueta (índice personalizado)
print("Edad de 'a':", serie_edades['a'])
print("Edad de 'c':", serie_edades['c'])
print()

# Acceder por posición (índice entero)
print("Primer elemento (posición 0):", serie_edades.iloc[0])
print("Tercer elemento (posición 2):", serie_edades.iloc[2])
print()

# Modificar el valor en el índice 'd'
print("Antes de modificar 'd':", serie_edades['d'])
serie_edades['d'] = 40
print("Después de modificar 'd':", serie_edades['d'])
print()

# Multiplicar toda la Series por 2
serie_doble = serie_edades * 2
print("Edades multiplicadas por 2:")
print(serie_doble)

print("\nSeries final (después de modificar 'd'):")
print(serie_edades)
```

# 2) DataFrame — crear y explorar


**Crea un DataFrame desde un diccionario y aplica un índice personalizado.**

```python
import pandas as pd

# 1. Crear DataFrame desde un diccionario
datos = {
    'nombre': ['Ana', 'Bob', 'Clara', 'Diego', 'Eva'],
    'edad': [25, 30, 22, 35, 28],
    'ciudad': ['Madrid', 'Lima', 'Bogotá', 'Medellín', 'Santiago'],
    'producto': ['Laptop', 'Teléfono', 'Tablet', 'Laptop', 'Tablet']
}

df = pd.DataFrame(datos)
print("DataFrame con índice por defecto (0, 1, 2...):")
print(df)
print()

# Definir un índice personalizado
indice_personalizado = ['venta_01', 'venta_02', 'venta_03', 'venta_04', 'venta_05']

# Asignar el nuevo índice
df.index = indice_personalizado

print("DataFrame con índice personalizado:")
print(df)
```

**Accede a una columna, usa loc para una fila por etiqueta y iloc por posición.**

```python
import pandas as pd

# Datos
datos = {
    'nombre': ['Ana', 'Bob', 'Clara', 'Diego', 'Eva'],
    'edad': [25, 30, 22, 35, 28],
    'ciudad': ['Madrid', 'Lima', 'Bogotá', 'Medellín', 'Santiago'],
    'producto': ['Laptop', 'Teléfono', 'Tablet', 'Laptop', 'Tablet']
}

# Índice personalizado
indice = ['venta_01', 'venta_02', 'venta_03', 'venta_04', 'venta_05']

# Crear DataFrame
df = pd.DataFrame(datos, index=indice)
print("DataFrame:")
print(df)
print()

# Acceder a la columna 'nombre'
columna_nombre = df['nombre']
print("Columna 'nombre':")
print(columna_nombre)
print()

# También puedes usar notación de punto (si el nombre no tiene espacios)
print("Columna 'edad' con df.edad:")
print(df.edad)
print()

# Acceder a la fila con etiqueta 'venta_03'
fila_loc = df.loc['venta_03']
print("Fila con loc['venta_03']:")
print(fila_loc)
print()

# Acceder a un valor específico: ciudad de 'venta_01'
valor_loc = df.loc['venta_01', 'ciudad']
print("Ciudad de venta_01 (loc):", valor_loc)

# Acceder a la fila en la posición 2 (tercera fila)
fila_iloc = df.iloc[2]
print("Fila en posición 2 (iloc[2]):")
print(fila_iloc)
print()

# Acceder a un valor por posición: nombre en fila 0, columna 0
valor_iloc = df.iloc[0, 0]
print("Nombre en posición [0,0] (iloc):", valor_iloc)

```

# 3) Operaciones básicas

**Agrega una columna derivada (por ejemplo, precio_descuento = precio * 0.9).**

```python
import pandas as pd

# DataFrame original del usuario
datos = {
    'nombre': ['Ana', 'Bob', 'Clara', 'Diego', 'Eva'],
    'edad': [25, 30, 22, None, 28],
    'ciudad': ['Madrid', 'Lima', 'Bogotá', 'Medellín', None],
    'producto': ['Laptop', 'Teléfono', 'Tablet', 'Laptop', 'Tablet'],
    'precio': [1200, 800, 300, 1150, None],
    'stock': [10, 15, 5, 8, 0]
}
df = pd.DataFrame(datos, index=['a', 'b', 'c', 'd', 'e'])

# Agregar columna con descuento del 10%
df['precio_descuento'] = df['precio'] * 0.9

print("DataFrame con columna 'precio_descuento' (10% de descuento):")
print(df)
```

**Aplica una operación vectorizada sobre una columna numérica.**

```python
import pandas as pd

# Tu DataFrame original (con columnas derivadas ya agregadas)
datos = {
    'nombre': ['Ana', 'Bob', 'Clara', 'Diego', 'Eva'],
    'edad': [25, 30, 22, None, 28],
    'ciudad': ['Madrid', 'Lima', 'Bogotá', 'Medellín', None],
    'producto': ['Laptop', 'Teléfono', 'Tablet', 'Laptop', 'Tablet'],
    'precio': [1200, 800, 300, 1150, None],
    'stock': [10, 15, 5, 8, 0]
}
df = pd.DataFrame(datos, index=['a', 'b', 'c', 'd', 'e'])

# Columnas derivadas (como hicimos antes)
df['precio_descuento'] = df['precio'] * 0.9
df['valor_inventario'] = df['precio'] * df['stock']
df['es_caro'] = df['precio'] > 900

# Operación vectorizada: multiplicar toda la columna 'stock' por 1.2
df['stock_aumentado'] = df['stock'] * 1.2

print("DataFrame con stock aumentado en 20%:")
print(df[['nombre', 'producto', 'stock', 'stock_aumentado']])
```

# 4) Manejo de datos faltantes

**Detecta nulos con isna y cuenta faltantes por columna.**

```python
import pandas as pd

# Tu DataFrame original (con columnas derivadas)
datos = {
    'nombre': ['Ana', 'Bob', 'Clara', 'Diego', 'Eva'],
    'edad': [25, 30, 22, None, 28],
    'ciudad': ['Madrid', 'Lima', 'Bogotá', 'Medellín', None],
    'producto': ['Laptop', 'Teléfono', 'Tablet', 'Laptop', 'Tablet'],
    'precio': [1200, 800, 300, 1150, None],
    'stock': [10, 15, 5, 8, 0]
}
df = pd.DataFrame(datos, index=['a', 'b', 'c', 'd', 'e'])

# Columnas derivadas (como antes)
df['precio_descuento'] = df['precio'] * 0.9
df['valor_inventario'] = df['precio'] * df['stock']
df['es_caro'] = df['precio'] > 900
df['stock_aumentado'] = df['stock'] * 1.2

# 1. Detectar valores nulos (devuelve True/False)
print("Detección de nulos con isna():")
print(df.isna())
print()

# 2. Contar nulos por columna
print("Número de valores faltantes por columna:")
print(df.isna().sum())
print()

# 3. Resumen general: ¿hay algún nulo en todo el DataFrame?
print("¿Hay algún valor nulo en todo el DataFrame?", df.isna().any().any())
print()

# 4. Total de valores nulos en todo el DataFrame
print("Total de valores nulos en el DataFrame:", df.isna().sum().sum())
```

**Completa faltantes en edad con 0 y en ciudad con 'Desconocido'.**

```python
import pandas as pd

# Tu DataFrame original (con columnas derivadas)
datos = {
    'nombre': ['Ana', 'Bob', 'Clara', 'Diego', 'Eva'],
    'edad': [25, 30, 22, None, 28],
    'ciudad': ['Madrid', 'Lima', 'Bogotá', 'Medellín', None],
    'producto': ['Laptop', 'Teléfono', 'Tablet', 'Laptop', 'Tablet'],
    'precio': [1200, 800, 300, 1150, None],
    'stock': [10, 15, 5, 8, 0]
}
df = pd.DataFrame(datos, index=['a', 'b', 'c', 'd', 'e'])

# Columnas derivadas (como antes)
df['precio_descuento'] = df['precio'] * 0.9
df['valor_inventario'] = df['precio'] * df['stock']
df['es_caro'] = df['precio'] > 900
df['stock_aumentado'] = df['stock'] * 1.2

# 1. Rellenar 'edad' con 0
df['edad'] = df['edad'].fillna(0)

# 2. Rellenar 'ciudad' con 'Desconocido'
df['ciudad'] = df['ciudad'].fillna('Desconocido')

print("DataFrame con valores faltantes completados:")
print(df)
```

**Ordena por precio descendente ignorando nulos (llena temporalmente nulos con 0 para ordenamiento).**

```python
import pandas as pd

# DataFrame original
datos = {
    'nombre': ['Ana', 'Bob', 'Clara', 'Diego', 'Eva'],
    'edad': [25, 30, 22, None, 28],
    'ciudad': ['Madrid', 'Lima', 'Bogotá', 'Medellín', None],
    'producto': ['Laptop', 'Teléfono', 'Tablet', 'Laptop', 'Tablet'],
    'precio': [1200, 800, 300, 1150, None],
    'stock': [10, 15, 5, 8, 0]
}
df = pd.DataFrame(datos, index=['a', 'b', 'c', 'd', 'e'])

# Ordenar por precio descendente, ignorando nulos (llenar temporalmente con 0)
df_ordenado = df.sort_values('precio', ascending=False, na_position='last').fillna({'precio': 0})
# O bien, ordenar con nulos al final y luego rellenar solo para mostrar
df_ordenado = df.sort_values('precio', ascending=False, na_position='last')
print(df_ordenado)
```

#9) Ejercicio integrado

**Aplica descuento del 10% a precio, filtra productos con stock > 5, ordena por precio_descuento y guarda como inventario_procesado.csv.**

```python
import pandas as pd

# Leer el CSV original
df = pd.read_csv('actividad_semana4.csv', index_col=0)

# 1. Aplicar descuento del 10% al precio
df['precio_descuento'] = df['precio'] * 0.9

# 2. Filtrar productos con stock > 5
df_filtrado = df[df['stock'] > 5]

# 3. Ordenar por precio_descuento (descendente)
df_ordenado = df_filtrado.sort_values('precio_descuento', ascending=False)

# 4. Guardar como nuevo CSV
df_ordenado.to_csv('inventario_procesado.csv', index=True)

print("Archivo 'inventario_procesado.csv' creado exitosamente.")
print(df_ordenado[['nombre', 'producto', 'precio', 'precio_descuento', 'stock']])
```
